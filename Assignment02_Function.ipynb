Q3)What are the different ways to define and call a function in Python?<br>
A3)Different Ways to Define and Call a Function in Python

1. Standard Function:
   - Defined using `def` keyword.
   - Example: `def greet(name): return f"Hello, {name}!"`
   - Call: `print(greet("Mayank"))`

2. Function with Default Arguments:
   - Allows default values for parameters.
   - Example: `def greet(name="Guest"): return f"Hello, {name}!"`
   - Call: `print(greet())` or `print(greet("Mayank"))`

3. Function with *args (Variable-Length Positional Arguments):
   - Allows multiple positional arguments.
   - Example: `def add_all(*numbers): return sum(numbers)`
   - Call: `print(add_all(1, 2, 3, 4, 5))`

4. Function with **kwargs (Variable-Length Keyword Arguments):
   - Allows multiple keyword arguments.
   - Example: `def display_info(**details): for key, value in details.items(): print(f"{key}: {value}")`
   - Call: `display_info(name="Mayank", age=25, city="London")`

5. Lambda (Anonymous) Function:
   - A small, one-line function.
   - Example: `square = lambda x: x * x`
   - Call: `print(square(5))`

6. Function as an Argument:
   - A function can be passed to another function.
   - Example: `def apply_function(func, value): return func(value)`
   - Call: `print(apply_function(double, 5))`

7. Function Inside Another Function (Nested Function):
   - A function defined inside another function.
   - Example: `def outer_function(name): def inner_function(): return f"Hello, {name}!" return inner_function()`
   - Call: `print(outer_function("Mayank"))`

8. Function with return Statement:
   - A function returning a value.
   - Example: `def multiply(a, b): return a * b`
   - Call: `result = multiply(4, 5); print(result)`

9. Recursion (Function Calling Itself):
   - A function calling itself.
   - Example: `def factorial(n): if n == 0 or n == 1: return 1 return n * factorial(n - 1)`
   - Call: `print(factorial(5))`


Q4)What is the purpose of the `return` statement in a Python function?<br>
A4) Purpose of the `return` Statement in a Python Function:
    - The `return` statement is used to send a value back from a function to the caller.
    - It allows functions to output results for further processing.
    - Example:<br>
      ```
      def square(num):
        return num * num
        result = square(4)
        print(result)  # Output: 16
      ```
    - If no `return` statement is used, the function returns `None` by default.

Q5) What are iterators in Python and how do they differ from iterables?<br>
A5)Iterators vs. Iterables:
    - **Iterable**: An object that can be iterated over (e.g., lists, tuples, dictionaries).
    - **Iterator**: An object that keeps state and produces values one at a time using `next()`.
    - Example:
      ```python
      my_list = [1, 2, 3]
      my_iter = iter(my_list)
      print(next(my_iter))  # Output: 1
      ```

Q6) Explain the concept of generators in Python and how they are defined.<br>
A6)Generators in Python:
    - Special functions using `yield` instead of `return`.
    - Generates values lazily, saving memory.
    - Example:
      ```python
      def count_up_to(n):
          count = 1
          while count <= n:
              yield count
              count += 1
      ```

Q7) What are the advantages of using generators over regular functions?<br>
A7)Advantages of Generators:
    - More memory-efficient than lists.
    - Used for large datasets or infinite sequences.
    - Example: `gen = count_up_to(3); print(next(gen))`

````python
#4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
````

````python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms  # Total terms in the sequence
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Track the number of terms generated

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.num_terms:  # Stop iteration when limit is reached
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            fib = self.a + self.b  # Compute next Fibonacci number
            self.a, self.b = self.b, fib  # Update values
            self.count += 1
            return fib

# Example: Generate first 10 Fibonacci numbers
fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num, end=" ")
````

````python
#6  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):  # Loop from 0 to the given exponent
        yield 2 ** i  # Yield the power of 2

# Example: Generate powers of 2 up to 2^5
for power in powers_of_2(5):
    print(power, end=" ")
````

````python
#7 Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    with open(file_path, "r") as file:
        for line in file:
            yield line.strip()  # Yield each line without trailing newline characters

# Example usage:
file_path = "example.txt"  # Replace with your file path

for line in read_file_line_by_line(file_path):
    print(line)  # Process each line one by one
````

````python
#8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
# List of tuples
data = [(1, 5), (3, 2), (4, 8), (2, 1)]

# Sorting using lambda function (based on the second element)
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
````

````python
#9  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Function to convert Celsius to Fahrenheit
celsius_to_fahrenheit = lambda c: (c * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Convert using map()
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print results
print(f"Temperatures in Fahrenheit: {fahrenheit_temperatures}")
````

````python
#10 Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Function to check if a character is not a vowel
remove_vowels = lambda char: char.lower() not in "aeiou"

# Input string
input_string = "Hello, World!"

# Use filter() to remove vowels
filtered_string = "".join(filter(remove_vowels, input_string))

# Print the result
print(f"String without vowels: {filtered_string}")

````
